```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import ta
import evaluation
import strategy
```

```python
bh_perf = pd.read_csv('evals/bh.csv', index_col='Date', parse_dates=['Date'])
macd_perf = pd.read_csv('evals/macd.csv', index_col='Date', parse_dates=['Date'])
lstm_perf = pd.read_csv('evals/lstm.csv', index_col='Date', parse_dates=['Date'])
ff_perf = pd.read_csv('evals/ff.csv', index_col='Date', parse_dates=['Date'])
```

```python
plt.figure(figsize=(25,12))

plt.plot(bh_perf['value'], label='Buy & Hold Strategy')
plt.plot(macd_perf['value'], label='MACD Stragety')
plt.plot(lstm_perf['value'], label='LSTM Stragety')
plt.plot(ff_perf['value'], label='FF Stragety')

plt.legend(loc='lower center', bbox_to_anchor=(0.15, 0.7), fontsize=22)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('Time', fontsize=28)
plt.ylabel('Portfolio Value', fontsize=28)

plt.show()
```

```python
plt.figure(figsize=(25,12))

plt.plot(bh_perf['percentage'], label='Buy & Hold Strategy')
plt.plot(macd_perf['percentage'], label='MACD Stragety')
plt.plot(lstm_perf['percentage'], label='LSTM Stragety')
plt.plot(ff_perf['percentage'], label='FF Stragety')

plt.legend(loc='lower center', bbox_to_anchor=(0.15, 0.7), fontsize=22)
plt.title('Performance of Different Strategies', fontsize=34)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('Time', fontsize=28)
plt.ylabel('Performance (% Increase)', fontsize=28)

plt.show()
```
