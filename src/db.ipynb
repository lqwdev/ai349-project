```python
import yfinance as yf
import matplotlib.pyplot as plt
from db import DailyStock
```

```python
db = DailyStock()
```

```python
tickers = [
    'AAPL',
    'FB',
    'MSFT',
    'GOOG',
    'AMZN',
    'QQQ',
    'DIA',
    'SPY',
]
```

```python
data = yf.download(
    tickers = ' '.join(tickers),
    start = "2010-01-01",
    end = "2021-01-01",
    interval = "1d",
    group_by = 'ticker',
    auto_adjust = True,
)
```

```python
data
```

```python
plt.figure(figsize=(17,8))

plt.plot(data['AAPL']['Close'], label='AAPL')
plt.plot(data['FB']['Close'], label='FB')
plt.plot(data['MSFT']['Close'], label='MSFT')
plt.plot(data['GOOG']['Close'], label='GOOG')
plt.plot(data['AMZN']['Close'], label='AMZN')

plt.plot(data['SPY']['Close'], label='SPY')
plt.plot(data['QQQ']['Close'], label='QQQ')
plt.plot(data['DIA']['Close'], label='DIA')

plt.legend(loc='lower center', bbox_to_anchor=(0.1, 0.6))

plt.show()
```

```python
for t in tickers:
    db.insert(t, data[t])
```

```python
df = db.daily('GOOG')
df
```

```python
plt.figure(figsize=(17,8))

plt.plot(df, label='GOOG')

plt.legend(loc='lower center', bbox_to_anchor=(0.1, 0.6))

plt.show()
```
