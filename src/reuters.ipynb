```python
from reuters import ReutersCrawler
from db import ReutersNews
import requests
```

```python
db = ReutersNews()
```

```python
crawler = ReutersCrawler()
```

```python
page = 2
while True:
    articles = crawler.fetch(page)
    db.insert(articles)
    page += 1
```
